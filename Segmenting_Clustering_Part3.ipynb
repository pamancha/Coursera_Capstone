```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml');
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
PostCodes1 = []
Borough1 = []
Neighborhood1 = []

for tr in My_table.find_all('tr')[2:]:
    tds = tr.find_all('td')
    PostCodes1.append(tds[0].text)
    Borough1.append(tds[1].text)
    Neighborhood1.append(tds[2].text)
```

```python
import pandas as pd
df1 = pd.DataFrame()
dfn = pd.DataFrame()
df1['PostalCode'] = PostCodes1
df1['Borough'] = Borough1


dfn['Neighborhood'] = Neighborhood1
dfn = dfn.replace('\n','', regex=True)
df1['Neighborhood'] = dfn

df1 = df1.drop(df1[df1.Borough == "Not assigned"].index)

df1.loc[df1['Neighborhood'] == "Not assigned", 'Neighborhood'] = df1['Borough']

df2 = df1
print(df2.head(100))
```

```python
df3 = df2.groupby(['PostalCode','Borough'])['Neighborhood'].apply(lambda x: "%s" % ', '.join(x))
df3=df3.reset_index()
#print(df3.head(10))
df3
```

### Calculating frequency of each category in neighbourhood

```python
df_freq = venues_onehot.groupby(['PostalCode' , 'Borough' , 'Neighborhood']).mean().reset_index()
print(df_freq.shape)
df_freq.head()
```

### Top 10 occurences of venue types in a neighbourhood

```python
num = 10
indicators = ['st' , 'nd' , 'rd']

area_col = ['PostalCode' , 'Borough' , 'Neighborhood']
freq_col = []
for gh in np.arange(num):
    try:
        freq_col.append('{}{} Most Common Venue'.format(gh+1 , indicators[ind]))
    except:
        freq_col.append('{}th Most Common Venue'.format(gh+1))
final_col = area_col + freq_col

df_sorted = pd.DataFrame(columns = final_col)
df_sorted['PostalCode'] = df_freq['PostalCode']
df_sorted['Borough'] = df_freq['Borough']
df_sorted['Neighborhood'] = df_freq['Neighborhood']

for gh in np.arange(df_freq.shape[0]):
    row_categories = df_freq.iloc[gh,:].iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending = False)
    df_sorted.iloc[gh , 3:] = row_categories_sorted.index.values[0:num]
    
df_sorted.sort_values(freq_col , inplace = True)
df_sorted
```
